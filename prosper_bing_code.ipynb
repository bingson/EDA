## Loan Data Cleaning and Reshaping

### Loan Data

```r
# rename state columns
colnames(state.gdp)  <- c("s.name","gdp")

# remove the 5 cancelled loan listings
raw <- raw %>% filter(LoanStatus != "Cancelled") 

# assign data types and factors
raw$LoanOriginationDate <- as.Date(raw$LoanOriginationDate, format = "%Y-%m-%d")
raw$LoanOriginationDate.year = format(raw$LoanOriginationDate, "%Y")
raw$ClosedDate          <- as.Date(substr(raw$ClosedDate, 1, 10), format = "%Y-%m-%d")
raw$LoanOriginalAmount  <- as.numeric(raw$LoanOriginalAmount) 
raw$OrigQtr <- substr(raw$LoanOriginationQuarter, 1,2) # separate quarter from year data
raw$OrigQtr <- factor(raw$OrigQtr)
raw$Term    <- factor(raw$Term)

# create dataframe of duplicates (excl. ProsperScore column) to investigate
dup.records <- raw[duplicated(raw[,c(1:15,17:83)]),] # omits first instance of duplicates
dup.ListingKey <- unique(c(as.character(dup.records$ListingKey)))

# includes first unique value and related duplicates
LD.duplicates <- raw %>% 
    filter(as.character(ListingKey) %in% dup.ListingKey) %>% 
    arrange(LoanOriginationDate)

# for plots that don't require ProsperScore, create subset of raw loan dataframe 
# that only has unique records after removing the ProsperScore column (column 16).
LD <- unique(raw[,c(1:15,17:83)])

# create loan data dataframe for credit risk analysis excluding all instances of ambiguous 
# ProsperScore records (all records with a duplicate).
# Omitted data represents around 1% of total lent dollar value and 
# around 0.7% of unique records
LD.prosperfix <- raw %>% 
    filter(!(as.character(ListingKey) %in% dup.ListingKey)) 

# gather credit risk related columns of interest for matrix analysis
credit <- LD.prosperfix[c(
    grep("^ListingKey", colnames(LD.prosperfix)),
    grep("^LoanOriginationDate", colnames(LD.prosperfix)),
    grep("^ClosedDate", colnames(LD.prosperfix)),
    grep("^Term", colnames(LD.prosperfix)),
    grep("^LoanOriginalAmount", colnames(LD.prosperfix)),
    grep("^LP_NetPrincipalLoss", colnames(LD.prosperfix)),
    grep("^CreditScore", colnames(LD.prosperfix)),
    grep("^CreditGrade", colnames(LD.prosperfix)),
    grep("^ProsperRating\\.\\.numeric", colnames(LD.prosperfix)),
    grep("^ProsperScore", colnames(LD.prosperfix))
)]

# Investigate rows with NAs before omission
# apply(c.risk, 1, function(r) any(r %in% c(NA))) gives logical vector of rows with NA
# NA.rows <- which(apply(credit, 1, function(r) any(r %in% c(NA))))
# NA.records <- credit[NA.rows,]

# add dummy variable indicating default events (charge offs + defaults) and loan loss %
# After noticing CreditGrade had 85125 NAs (over half the dataset), 
# decided to remove CreditGrade from further consideration.
# prosper1.0 loans were also ommited (pre 2009 issuance)
# as well as entries that were missing one of the credit rating related columns
credit <- credit %>%
    select(-CreditGrade) %>% 
    mutate(lossPct = LP_NetPrincipalLoss/LoanOriginalAmount,
           default = as.factor(ifelse(lossPct > 0, 1 ,0))) %>% 
    filter(LoanOriginationDate.year > 2008, 
           !is.na(ProsperScore), 
           !is.na(CreditScoreRangeUpper),
           !is.na(ProsperRating..numeric.)) %>% 
    arrange(LoanOriginationDate)

# join CreditScore upper and lower ranges to create FICO borrower buckets
credit$fico <- do.call(paste, c(credit[c("CreditScoreRangeLower", 
                                         "CreditScoreRangeUpper")], sep = "-")) 

# convert credit score to discrete scale numbering 1-15 to simplify presentation 
# and communicate relative credit rating bucket position. 
credit$fico <- factor(credit$fico)
credit$f.scr <- as.integer(credit$fico)

# Rename Prosper score and Prosper rating column to p.scr and p.rtg
credit <- credit %>% 
    select(-CreditScoreRangeLower:-ProsperScore,
           p.rtg = ProsperRating..numeric.,
           p.scr = ProsperScore)

# subset loans that were closed (i.e, status = completed, defaulted, charged-off)
credit.closed <- credit %>% filter(!is.na(ClosedDate))
```

### Base Rates

```r
# change base market rate dataframes to appropriate data types and fill 
clean_rate_data <- function(src_data){  
    src_data %>% mutate(date = as.Date(date, format = "%Y-%m-%d"),
                        rate = as.numeric(as.character(rate)), 
                        rate = rate/100) %>% 
    transform(rate = na.locf(rate)) # carry fwd last rate if NA
    }

BM12mr <- clean_rate_data(raw12mr)
BM36mr <- clean_rate_data(raw36mr)
BM60mr <- clean_rate_data(raw60mr)

rm(raw12mr, raw36mr, raw60mr, raw)
```

```r
LD.prosperfix <- LD.prosperfix %>% 
    select(ListingKey,
           year = LoanOriginationDate.year, 
           date = LoanOriginationDate, 
           dollars = LoanOriginalAmount,
           Term,
           credit.rating = ProsperRating..Alpha.,
           EstimatedEffectiveYield,
           EstimatedLoss,
           EstimatedReturn,
           BorrowerAPR,
           BorrowerRate,
           LenderYield) 

br12 <- LD.prosperfix %>%
    filter(year > "2008", Term == "12") %>% 
    arrange(date) %>% 
    left_join(BM12mr, by = "date") %>% 
    select(ListingKey, year:LenderYield, rate)

br36 <- LD.prosperfix %>%
    filter(year > "2008", Term == "36") %>% 
    arrange(date) %>% 
    left_join(BM12mr, by = "date") %>% 
    select(ListingKey, year:LenderYield, rate)

br60 <- LD.prosperfix %>%
    filter(year > "2008", Term == "60") %>% 
    arrange(date) %>% 
    left_join(BM12mr, by = "date") %>% 
    select(ListingKey, year:LenderYield, rate)
```

```r
rates_combined <- rbind(br12, br36, br60)

Spread <- rates_combined %>%
    filter(credit.rating != "", !is.na(rate)) %>% 
    arrange(date) %>% 
    mutate(riskSpread = BorrowerRate - rate)

Spread$credit.rating <- ordered(Spread$credit.rating, levels = pr.lab)

Spread <- Spread %>% left_join(credit %>% select(ListingKey, 
                                                 f.scr, 
                                                 fico, 
                                                 p.scr), by = "ListingKey")

```

## Loan Defaults and Durations

### Data Reshaping

```r
# merging new loan and write-offs (net of recoveries) to calculate default rates-----
# Loan origination amounts by vintage
OrigVintage <- LD %>%
    arrange(LoanOriginationDate) %>%
    select(OrigYear = LoanOriginationDate.year,
           Date = LoanOriginationDate, 
           newLoanAmt = LoanOriginalAmount)
OrigVintage$NetLoss <- 0
OrigVintage$newLoanAmt <- as.double(OrigVintage$newLoanAmt)

# Defaults and Charge-offs, net of recoveries
DefaultVintage <- LD %>%
    arrange(ClosedDate) %>% 
    select(OrigYear = LoanOriginationDate.year, 
           Date = ClosedDate, 
           NetLoss = LP_NetPrincipalLoss)
DefaultVintage$newLoanAmt <- 0

# combining new loan issuance and net write-offs in chronological order (wide -> long)
# to calculate default rates over time by loan vintage
D <- OrigVintage %>% rbind(DefaultVintage) %>% 
    filter(OrigYear > 2005)

D <- D %>%
    arrange(Date) %>% 
    group_by(OrigYear) %>% 
    mutate(NewLoanBal = cumsum(newLoanAmt), 
           NetLossBal = cumsum(NetLoss), 
           cum.loss.pct = NetLossBal/NewLoanBal*100) %>% 
    filter(!is.na(Date))

cumloss <- D %>% 
    group_by(OrigYear) %>% 
    filter(row_number(desc(cum.loss.pct)) == 1) %>% 
    select(cum.loss.pct) 
 
closed.loans <- LD %>% 
    group_by(LoanOriginationDate.year) %>% 
    filter(!is.na(ClosedDate)) %>% 
    summarise(closedLoans = n(), closedDollars = sum(LoanOriginalAmount)) 
  
total.loans <- LD %>% 
    group_by(LoanOriginationDate.year) %>% 
    summarise(totalLoans = n(), totalDollars = sum(LoanOriginalAmount)) 

# create table summarizing annual loan default characteristics
default.table <- cumloss %>% 
    left_join(closed.loans, by = c("OrigYear" = "LoanOriginationDate.year")) %>% 
    left_join(total.loans, by = c("OrigYear" = "LoanOriginationDate.year")) %>% 
    mutate(cum.loss.pct = round(cum.loss.pct, digits = 1), 
           loans.closed.pct = round(closedLoans/totalLoans*100, digits = 1),
           dollars.closed.pct = round(closedDollars/totalDollars*100, digits = 1),
           p.o = (totalDollars-closedDollars)) %>%
    select(OrigYear, 
           'dollars.raised' = totalDollars,
           'dollars.outstanding' = p.o, 
           'pct.matured' = dollars.closed.pct,  # includes defaulted
           'pct.charged.off' = cum.loss.pct) %>% 
    arrange(OrigYear)

D <- left_join(D, default.table[,c("OrigYear","dollars.raised")]) 


# The below calculations are related to finding cumulative defaults rates on 2011 and 2012
# at the equivalent point in the maturity cycle as current 2013 vintage loans. 
# In other words, according to default.table, for 2013 originated loans 6.5% of 
# loans have so far been closed (matured or charged-off) with a cumulative default 
# rate of 0.7%. Below I find out what the cumulative default rate on 2011 and 2012 
# vintages when 6.5% of total dollars lent in those vintages had been closed.

mat.2013pct <- default.table$pct.matured[default.table$OrigYear == 2013]/100
mat.2013ddpct <- default.table$pct.charged.off[default.table$OrigYear == 2013]

dollars.2011  <- default.table$dollars.raised[default.table$OrigYear == 2011]
dollars.2012  <- default.table$dollars.raised[default.table$OrigYear == 2012]

# find the date for 2011 vintage loans when at 6.5% closed 
date.2013mpct.2011 <- LD %>%
    filter(!is.na(ClosedDate), LoanOriginationDate.year == "2011") %>% 
    arrange(ClosedDate) %>%
    mutate(pct.matured = cumsum(LoanOriginalAmount)/dollars.2011) %>% 
    filter(pct.matured <= mat.2013pct) %>% 
    select(ClosedDate, pct.matured) %>% 
    filter(row_number(desc(pct.matured)) == 1)

# find the date for 2012 vintage loans when at 6.5% closed
date.2013mpct.2012 <- LD %>%
    filter(!is.na(ClosedDate), LoanOriginationDate.year == "2012") %>% 
    arrange(ClosedDate) %>%
    mutate(pct.matured = cumsum(LoanOriginalAmount)/dollars.2012) %>% 
    filter(pct.matured <= mat.2013pct) %>% 
    select(ClosedDate, pct.matured) %>% 
    filter(row_number(desc(pct.matured)) == 1)

d.2012.matpct.2013 <- D %>% filter(OrigYear == "2012", 
                                   Date == date.2013mpct.2012[1,1]) %>% 
    summarise(median(cum.loss.pct))

d.2011.matpct.2013 <- D %>% filter(OrigYear == "2011", 
                                   Date == date.2013mpct.2011[1,1]) %>%
    summarise(median(cum.loss.pct))
```

### Figure 17: Credit Rating Distributions 2013 

```r

# create mean, median, interquartile range lines
mcuts <- function(col, wt, guide = FALSE) {
    mn <- weighted.mean(col, w = wt) 
    md <- weighted.median(col, w = wt) 
    mn.c <- data.frame(midpoints="dollar-weighted mean", vals = mn)
    md.c <- data.frame(midpoints="dollar-weighted median", vals = md)
    MM <- rbind(mn.c, md.c)
  
    q1 <- wtd.quantile(col, q=0.25, na.rm = FALSE, weight = wt)
    q3 <- wtd.quantile(col, q=0.75, na.rm = FALSE, weight = wt)
    q1.c <- data.frame(quantile = "1st quartile", vals = q1)
    q3.c <- data.frame(quantile = "3rd quartile", vals = q3)
    QQ <- rbind(q1.c, q3.c)
  
    list(
        geom_vline(data=MM,
                   aes(xintercept = vals, 
                       linetype = midpoints),
                   show_guide = guide, 
                   size = 1.1),
        guides(fill=guide_legend(title=NULL)),
        geom_vline(data=QQ,
                   aes(xintercept = vals),
                   show_guide = FALSE, 
                   linetype = 1,
                   size = 1,
                   color = afill)
  
  )
}

# reshape FICO data

# merge high risk credit buckets to simplify presentation
credit.f <- credit %>% 
    mutate(fico.f = ifelse(fico == "600-619" | fico == "620-639" | fico == "640-659",
                           "600-659", as.character(fico)), 
           f.scr.f = as.numeric(factor(fico.f)))

fico.2013q123 <- credit.f %>%
    filter(LoanOriginationDate >= "2013-01-01", 
           LoanOriginationDate < "2013-10-01") %>%
    group_by(f.scr.f) %>% 
    summarise(dollars = sum(LoanOriginalAmount), loans = n()) %>% 
    mutate(wt = dollars/sum(credit$LoanOriginalAmount)) %>% 
    select(score = f.scr.f, dollars, loans, wt)

fico.2013q4 <- credit.f %>%
    filter(LoanOriginationDate >= "2013-10-01", 
           LoanOriginationDate < "2014-01-01") %>% 
    group_by(f.scr.f) %>% 
    summarise(dollars = sum(LoanOriginalAmount), loans = n()) %>% 
    mutate(wt = dollars/sum(credit$LoanOriginalAmount)) %>% 
    select(score = f.scr.f, dollars, loans, wt) %>% 
    # create place holders for empty buckets
    rbind(data.frame(score = 12, dollars = 0, loans = 0, wt = 0)) %>% 
    rbind(data.frame(score = 13, dollars = 0, loans = 0, wt = 0))     
    # no issuance from FICO 860-899 in 2013Q4

# Prosper Ratings reshape
pr.2013q123 <- credit %>%
    filter(LoanOriginationDate >= "2013-01-01", 
           LoanOriginationDate < "2013-10-01") %>% 
    group_by(p.rtg) %>% 
    summarise(dollars = sum(LoanOriginalAmount), loans = n()) %>% 
    mutate(wt = dollars/sum(credit$LoanOriginalAmount)) %>% 
    select(score = p.rtg, dollars, loans, wt)

pr.2013q4 <- credit %>%
    filter(LoanOriginationDate >= "2013-10-01", 
           LoanOriginationDate < "2014-01-01") %>%
    group_by(p.rtg) %>% 
    summarise(dollars = sum(LoanOriginalAmount), loans = n()) %>% 
    mutate(wt = dollars/sum(credit$LoanOriginalAmount)) %>% 
    select(score = p.rtg, dollars, loans, wt)

# create plots

# fine tune plot margins
# c(top, right, bottom, left)
fq123 <- unit(c( 0.3, -0.5,  0   ,  0  ), "cm")
fq4   <- unit(c( 0  , -0.5,  0.15,  0  ), "cm")
pq123 <- unit(c( 0.3,  0,    0   , -0.5), "cm")
pq4   <- unit(c( 0  ,  0,    1   , -0.5), "cm")

P.f.2013q123 <- ggplot(data = fico.2013q123, aes(x=score, y =dollars/units)) + 
    theme2 + geom_bar(stat = "identity", fill = afill) +
    ggtitle("FICO Score\n2013 Q1 - Q3") +
    scale_y_continuous(breaks = seq(0,100,25), limits = c(0, 100)) +
    scale_x_continuous(breaks = seq(1, 13, 1)) +
    ylab("Dollars (millions)") + 
    mcuts(fico.2013q123$score, fico.2013q123$dollars, T) +
    annotate("text", x = 7, y = 50,
             label = "interquartile\nrange", 
             hjust = 0, color = "gray55", size = 5) +
    guides(linetype = guide_legend(override.aes = list(colour = NULL), title=NULL)) +
    theme(legend.position=c(0.68,0.74), 
          legend.direction="vertical", 
          plot.margin = fq123)

P.f.2013q4 <- ggplot(data = fico.2013q4, aes(x=score, y =dollars/units)) + 
    theme4 + geom_bar(stat = "identity", fill = afill) +
    xlab("FICO Score\n2013 Q4") +
    scale_y_continuous(breaks = seq(0,100,25), limits = c(0, 100)) +
    scale_x_continuous(breaks = seq(1, 13, 1), 
                       labels = levels(factor(unique(credit.f$fico.f)))) +
    ylab("Dollars (millions)") + 
    mcuts(fico.2013q4$score, fico.2013q4$dollars) +
    theme(plot.margin = fq4,
          axis.title.x = element_text(vjust = 0),
          axis.text.x = element_text(angle = 45, size = 10, hjust = 1, vjust = .9))


p.pr.2013q123 <- ggplot(data = pr.2013q123, aes(x=score, y =dollars/units)) + theme2 +
    geom_bar(stat = "identity", width = 7/13, fill = afill) +
    ggtitle("Prosper Rating\n2013 Q1 - Q3") +
    scale_y_continuous(breaks = seq(0,100,25), limits = c(0, 100)) +
    scale_x_continuous(breaks = seq(1,7,1)) + 
    mcuts(pr.2013q123$score, pr.2013q123$dollars) +
    theme(axis.title.y = element_blank(), 
          axis.text.y = element_blank(),
          axis.ticks = element_blank(),
          plot.margin = pq123)

p.pr.2013q4 <- ggplot(data = pr.2013q4, aes(x=score, y =dollars/units)) + theme4 +
    geom_bar(stat = "identity", width = 7/13, fill = afill) +
    xlab("Prosper Rating\n2013 Q4")+
    scale_y_continuous(breaks = seq(0,100,25), limits = c(0, 100)) +
    scale_x_continuous(breaks = seq(1,7,1), labels = rev(pr.lab)) + 
    mcuts(pr.2013q4$score, pr.2013q4$dollars) +
    theme(axis.title.y = element_blank(), 
          axis.title.x=element_text(vjust= -3.1),
          axis.text.y = element_blank(),
          axis.ticks = element_blank(),
          plot.margin = pq4)

# draw plots
grid.arrange(P.f.2013q123,p.pr.2013q123,
             P.f.2013q4, p.pr.2013q4,
             ncol = 2)
```
